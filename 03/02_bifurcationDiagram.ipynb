# 分岐図 ifバージョン

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
K = 100

rList = []
xList = []

for i in range(150,300):
    x0 = 10
    r = i/100
    
    x=x0
    for t in range(1000):
        xx =x + r* (1-x/K)*x
        x = xx
        if t>900:
            rList.append(r)
            xList.append(x)
```

```python
plt.plot(rList, xList, 'r+', markersize="1")
```

```python
K = 100

rList = []
xList = []

for i in range(150,300):
    x0 = 10
    r = i/100
    
    x=x0
    for t in range(900):
        xx =x + r* (1-x/K)*x
        x = xx
    for t in range(100):
        xx =x + r* (1-x/K)*x
        x = xx
        rList.append(r)
        xList.append(x)
```

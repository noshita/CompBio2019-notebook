# matplotlibによる作図

```python
# ノートブック内に図を出力するためのコマンド
%matplotlib inline
import matplotlib.pyplot as plt # matplotlib.pyplotをpltとして読み込む
import math # mathライブラリを読み込む（sin関数を使うため）
```

```python
xEnd = 2*math.pi
step = math.pi/4 # プロット時の刻み幅

xList = [] # 空のリストを作成
yList = [] # 空のリストを作成
for i in range(0, int(xEnd/step)):
    x = step*i
    y = math.sin(x)
    # xとyの座標値をそれぞれリストの末尾に追加
    xList.append(x)
    yList.append(y)
```

```python
# 作図
plt.plot(xList,yList)
```
